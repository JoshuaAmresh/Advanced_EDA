### Import Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use("seaborn-darkgrid")
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
sns.set(style="white")
```

### Read the data

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

### Regression Plot

Regression plot creates a regression line between 2 parameters and helps to visualize their linear relationships

```python
sns.regplot(x=df.LotArea, y=df.SalePrice, color='g')
```

```python
sns.regplot(x=df.LotFrontage, y=df.SalePrice, color='g')
```

```python
sns.regplot(x=df.MSSubClass, y=df.SalePrice, color='g')
```

### Bivariate analysis for independent variable being a category and dependent variable being a number

```python
sns.barplot(x=df.MSSubClass, y=df.SalePrice)
```

Mean value for each MSSubClass value

```python
sns.barplot(x=df.LotShape, y=df.SalePrice)
```
